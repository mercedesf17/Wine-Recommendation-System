```python
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
import numpy as np
import random
```

```python
geolocation_df = pd.read_csv('data/latlon_wine_data.csv')
```

```python
expanded_df = pd.read_csv('data/expanded_dataframe.csv')
expanded_df['address'] = expanded_df['region'] + ', ' + expanded_df['province'] + ', ' + expanded_df['country']
```

```python
expanded_df= expanded_df.rename(columns={'Unnamed: 0': 'index'})
```

```python
expanded_df.columns
```

```python
south_eu = ["Italy", "Spain", "Portugal", 'Greece']
south_af =  ["South Africa"]
oc = ["New Zealand", "Australia"]
south_am= ["Chile", "Argentina"]
north_am= ["United States", "Canada"]
central_eu = ["France","Germany"]
israel = ["Israel"]
```

```python
country_dictionary = {"Portugal": 1, "Spain": 2, "France": 3, "Germany": 4, "Austria": 5,
                      "Italy": 6, "Greece": 7, "Israel": 8, "South Africa": 9, "Australia": 10,
                      "New Zealand": 11, "Chile": 12, "Argentina": 13, "US": 14, "Canada": 15}

expanded_df["encoded_country"] = expanded_df['country'].map(country_dictionary)
```

```python
price_input = (10,30)
wine_type_iput = 'red'
```

```python
filtered_df = expanded_df[(expanded_df['price'] >= price_input[0]) & (expanded_df['price'] <= price_input[1]) & (expanded_df['wine_type'] == wine_type_iput)]
filtered_df
```

```python
# Fit the KNN model

X = filtered_df[['dry_wine', 'sweet_wine', 'fruity_aroma','spicy_aroma', 'herb_aroma', 'oak_aroma', 'chocolate_aroma','floral_aroma', 'body_light', 'body_medium', 'body_full', 'encoded_country']]
y = filtered_df['index']

neigh = KNeighborsRegressor(n_neighbors=15)
neigh.fit(X, y)
```

```python
X_predict = np.array([[0, 1, 1, 0, 0, 0, 0, 1,0, 0, 0,5]])

neigh.kneighbors(X_predict)[1]
```
